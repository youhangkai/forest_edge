```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def draw_sierpinski_triangle(ax, vertices, depth):
    if depth == 0:
        # Draw the triangle
        ax.fill(*zip(*vertices), edgecolor='black', fill=False)
    else:
        # Calculate the mid-points of each side
        p1 = (vertices[0] + vertices[1]) / 2
        p2 = (vertices[1] + vertices[2]) / 2
        p3 = (vertices[0] + vertices[2]) / 2

        # Recursively draw the three smaller triangles
        draw_sierpinski_triangle(ax, [vertices[0], p1, p3], depth-1)
        draw_sierpinski_triangle(ax, [p1, vertices[1], p2], depth-1)
        draw_sierpinski_triangle(ax, [p3, p2, vertices[2]], depth-1)

# Create a plot
fig, axes = plt.subplots(1, 5, figsize=(20, 4))

# Initial vertices of the triangle
initial_vertices = np.array([[0, 0], [1, 0], [0.5, np.sqrt(3)/2]])

# Draw Sierpinski triangles for n = 1 to 5
for i in range(5):
    draw_sierpinski_triangle(axes[i], initial_vertices, i+1)
    axes[i].set_xlim(-0.1, 1.1)
    axes[i].set_ylim(-0.1, 1.1)
    axes[i].set_aspect('equal')
    axes[i].axis('off')

plt.show()
```
