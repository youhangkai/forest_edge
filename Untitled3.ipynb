```python
import rasterio
import matplotlib.pyplot as plt

# 打开图像文件
image_file = r'H:\Global_tree_cover\2000extent\00N_000E.tif'  # 替换为你的图像文件路径
image = rasterio.open(image_file)

# 读取图像数据
image_data = image.read(1)  # 读取第一个波段的数据

# 显示图像
plt.figure(figsize=(8, 8))
plt.imshow(image_data, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# 转置图像
transposed_image_data = image_data.T

# 显示图像
plt.figure(figsize=(8, 8))
plt.imshow(transposed_image_data, cmap='gray')
plt.axis('off')
plt.show()
```
