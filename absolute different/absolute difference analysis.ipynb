```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt

# Path to your TIFF file
tif_path = "G:/Hangkai/Forest_edge_mapping_data/Global_001_degree/2020Edge_2000Edge_diff.tif"

# Read the TIFF file
with rasterio.open(tif_path) as src:
    edge_data = src.read(1)  # Read the first band

    # Mask out no-data values and zeros
    nodata = src.nodata
    if nodata is not None:
        edge_data = np.ma.masked_equal(edge_data, nodata)
    edge_data = np.ma.masked_equal(edge_data, 0)
    edge_data = edge_data/(1000)

    # Calculating statistics
    max_value = np.max(edge_data)
    min_value = np.min(edge_data)
    
    percentile_95 = np.percentile(edge_data.compressed(), 95)
    percentile_5 = np.percentile(edge_data.compressed(), 5)

    # Print out the statistics
    print(f"Maximum Value: {max_value} km")
    print(f"Minimum Value: {min_value} km")
    print(f"95th Percentile: {percentile_95} km")
    print(f"5th Percentile: {percentile_5} km")

    # Plotting the distribution
    plt.figure(figsize=(10, 6))
    plt.hist(edge_data.compressed(), bins=50, color='blue', alpha=0.7)
    plt.title("Distribution of Forest Edge Length Differences (2000-2020) Excluding Zeros")
    plt.xlabel("Difference in Length")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()
```

```python
for i in range(0,6):
    percentile_value = np.percentile(edge_data.compressed(), i)
    print(f"{i}th Percentile: {percentile_value} km")
for i in range(95,101):
    percentile_value = np.percentile(edge_data.compressed(), i)
    print(f"{i}th Percentile: {percentile_value} km")

```

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt

# Path to your TIFF file
tif_path = "G:/Hangkai/Forest_edge_mapping_data/Global_001_degree/2020Area_2000Area_diff.tif"

# Read the TIFF file
with rasterio.open(tif_path) as src:
    area_data = src.read(1)  # Read the first band

    # Mask out no-data values if necessary
    nodata = src.nodata
    if nodata is not None:
        area_data = np.ma.masked_equal(area_data, nodata)
    area_data = np.ma.masked_equal(area_data, 0)
    area_data = area_data/(1000*1000)

    # Calculating statistics
    max_value = np.max(area_data)
    min_value = np.min(area_data)
    percentile_95 = np.percentile(area_data.compressed(), 95)
    percentile_5 = np.percentile(area_data.compressed(), 5)

    # Print out the statistics
    print(f"Maximum Value: {max_value} km2")
    print(f"Minimum Value: {min_value} km2")
    print(f"95th Percentile: {percentile_95} km2")
    print(f"5th Percentile: {percentile_5} km2")

    # Plotting the distribution
    plt.figure(figsize=(10, 6))
    plt.hist(area_data.flatten(), bins=50, color='red', alpha=0.7)
    plt.title("Distribution of Forest Area Differences (2000-2020)")
    plt.xlabel("Difference in Area")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()
```

```python
for i in range(0,6):
    percentile_value = np.percentile(area_data.compressed(), i)
    print(f"{i}th Percentile: {percentile_value} km2")
for i in range(95,101):
    percentile_value = np.percentile(area_data.compressed(), i)
    print(f"{i}th Percentile: {percentile_value} km2")

```
