```python
import numpy as np
import time
from global_land_mask import globe
import matplotlib.pyplot as plt

# Lat/lon points to get
lat = np.linspace(-60, 80, 14000)
lon = np.linspace(-180, 180, 36000)

# Make a grid
lon_grid, lat_grid = np.meshgrid(lon, lat)

# Get whether the points are on land using globe.is_land
start_time = time.time()
globe_land_mask = globe.is_land(lat_grid, lon_grid)
globe_run_time = time.time() - start_time
print('Time to run globe.is_land(): {}'.format(globe_run_time))

# Check agreement (note there is a different treatment for lakes)
fraction_agreed = np.sum(globe_land_mask) / len(globe_land_mask.flatten())
print('Fraction on land: {}'.format(fraction_agreed))

# Plot the land mask as a raster
plt.figure(figsize=(10, 6))
plt.imshow(globe_land_mask, extent=[lon.min(), lon.max(), lat.min(), lat.max()], cmap='binary', origin='lower')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Global Land Mask')
plt.colorbar(label='Land')
plt.grid(True)
plt.show()
```
