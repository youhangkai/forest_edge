```python
import pandas as pd

# Load the data
file_path = 'G:/Hangkai/Forest_edge/country_stats/final_data/Edge_climate_zone_stats.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
data.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Setting up the figure and axes
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# Climate Zone Comparison: Total edges in 2000 and 2020
climate_zones = data['Climate Zone']
edges_2000 = data['2000 Edge']
edges_2020 = data['2020 Edge']

x = np.arange(len(climate_zones))  # the label locations

# Plotting
width = 0.35  # the width of the bars
rects1 = axs[0].bar(x - width/2, edges_2000, width, label='2000')
rects2 = axs[0].bar(x + width/2, edges_2020, width, label='2020')

# Adding labels, title, and custom x-axis and y-axis tick labels, etc.
axs[0].set_ylabel('Edges')
axs[0].set_title('Total edges by climate zone and year')
axs[0].set_xticks(x)
axs[0].set_xticklabels(climate_zones, rotation=45, ha='right')
axs[0].legend()

# Climate Zone Comparison: Stable edges in 2000 and 2020
stable_edges = data['stable']

# Plotting
rects1 = axs[1].bar(x - width/2, stable_edges, width, label='2000', alpha=0.7)
rects2 = axs[1].bar(x + width/2, stable_edges, width, label='2020', alpha=0.7)

# Adding labels, title, and custom x-axis and y-axis tick labels, etc.
axs[1].set_ylabel('Stable Edges')
axs[1].set_title('Stable edges by climate zone')
axs[1].set_xticks(x)
axs[1].set_xticklabels(climate_zones, rotation=45, ha='right')
axs[1].legend()

# Auto format the date on the x-axis and make the layout tight
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the climate zones
climate_zones = data['Climate Zone']

# Define the edge data for 2000 and 2020
edge_2000 = data['2000 Edge']
edge_2020 = data['2020 Edge']

# Define the percentages
edge_2000_per = data['2000 Edge Per'] * 100  # converting to percentage
edge_2020_per = data['2020 Edge Per'] * 100  # converting to percentage

x = np.arange(len(climate_zones))  # the label locations

fig, ax = plt.subplots(figsize=(10, 6))

# Plotting
bar_width = 0.35  # the width of the bars
rects1 = ax.bar(x - bar_width/2, edge_2000, bar_width, label='2000 Edge', alpha=0.7)
rects2 = ax.bar(x + bar_width/2, edge_2020, bar_width, label='2020 Edge', alpha=0.7)

# Adding percentages on top of the bars
for i, rect in enumerate(rects1):
    height = rect.get_height()
    ax.annotate(f'{edge_2000_per[i]:.1f}%',
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

for i, rect in enumerate(rects2):
    height = rect.get_height()
    ax.annotate(f'{edge_2020_per[i]:.1f}%',
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

# Adding labels, title, and custom x-axis and y-axis tick labels, etc.
ax.set_xlabel('Climate Zone')
ax.set_ylabel('Edge (km)')
ax.set_title('Edge by Climate Zone and Year')
ax.set_xticks(x)
ax.set_xticklabels(climate_zones, rotation=45, ha='right')
ax.legend()

fig.tight_layout()

plt.show()
```

```python
# Creating a bar plot for each climate zone comparing the total edges in 2000 and 2020
num_zones = len(climate_zones)

fig, axs = plt.subplots(2, 3, figsize=(8,4), sharex=True, tight_layout=True)

for i, zone in enumerate(climate_zones):
    
    axs[(i%2),(i%3)].bar('2000', edge_2000[i], color='blue', alpha=0.7, label='2000 Edge')
    axs[(i%2),(i%3)].bar('2020', edge_2020[i], color='red', alpha=0.7, label='2020 Edge')
    axs[(i%2),(i%3)].set_ylabel('Edge (km)')
    axs[(i%2),(i%3)].set_title(f'Comparison: {zone}')
    
axs[0,0].legend()
plt.xlabel('Year')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to plot the edge types for each climate zone
def plot_edge_types(data, edge_types, title):
    # Number of climate zones
    num_zones = len(data)
    
    # Colors for the different edge types
    colors = ['red', 'green', 'blue', 'purple', 'orange']
    
    # Calculate the total for percentage calculation
    total = data[edge_types].sum(axis=1)
    
    # Plotting
    plt.figure(figsize=(10, 6))
    for i, edge_type in enumerate(edge_types):
        percentages = (data[edge_type] / total) * 100
        plt.bar(data['Climate Zone'], data[edge_type], bottom=data[edge_types[:i]].sum(axis=1),
                color=colors[i], label=f'{edge_type} ({percentages.mean():.2f}%)')
        
        # Adding text on the bars
        cum_values = data[edge_types[:i]].sum(axis=1)
        for j, (value, cum_value) in enumerate(zip(data[edge_type], cum_values)):
            if value > 0:
                plt.text(j, cum_value + value/2, f'{percentages.iloc[j]:.2f}%', ha='center', va='center', color='white')
    
    # Adding labels and titles
    plt.xlabel('Climate Zone')
    plt.ylabel('Edge Count')
    plt.title(title)
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Edge types to be plotted
edge_types = ['stable', 'ii', 'id', 'di', 'dd']

# Plotting the edge types for each climate zone
plot_edge_types(data, edge_types, 'Edge Types by Climate Zone (2000 and 2020)')
```

```python
# 2020 Edge values
edge_2020_values = [60697800, 52315750, 2286745.5, 62031996, 70308170, 46052884]

# Plotting stacked bar chart with 2020 Edge values as heights and percentages on each section
plt.figure(figsize=(12, 7))

# Defining a function to add percentage text
def add_text(bar, value, total):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_y() + height / 2,
             f'{(height/total)*100:.1f}%', ha='center', va='center', color='white', fontsize=10)

# Plotting each column
bars_ii = plt.bar(data['Climate Zone'], data['ii/2020 per'] * edge_2020_values, color='#1f77b4', label='ii/2020 per')
bars_id = plt.bar(data['Climate Zone'], data['id/2020 per'] * edge_2020_values, 
                  bottom=data['ii/2020 per'] * edge_2020_values, color='#ff7f0e', label='id/2020 per')
bars_stable = plt.bar(data['Climate Zone'], data['Stable 2020 per'] * edge_2020_values, 
                      bottom=(data['ii/2020 per'] + data['id/2020 per']) * edge_2020_values, color='#2ca02c', label='Stable 2020 per')

# Adding percentages on the bars
for bar, value, total in zip(bars_ii, data['ii/2020 per'] * edge_2020_values, edge_2020_values):
    add_text(bar, value, total)
for bar, value, total in zip(bars_id, data['id/2020 per'] * edge_2020_values, edge_2020_values):
    add_text(bar, value, total)
for bar, value, total in zip(bars_stable, data['Stable 2020 per'] * edge_2020_values, edge_2020_values):
    add_text(bar, value, total)

# Adding labels and titles
plt.xlabel('Climate Zone')
plt.ylabel('Edge Count (km)')
plt.title('Edge Types by Climate Zone (2020)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
