```python
import numpy as np
import matplotlib.pyplot as plt
import rasterio
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy
import matplotlib.colors as colors
import matplotlib.cm as cm

# Set the file path
file_path = r'H:\Global_tree_cover\global_raster\20_00_diff.tif'

# Open the file using rasterio
dataset = rasterio.open(file_path)

# Read the data
data = dataset.read(1)

# Get the geospatial transformation information
transform = dataset.transform

# Get the geographic bounds
bounds = dataset.bounds

# Create a new figure and axis
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

# Plot the data with specified vmin and vmax
data = data * 100
# data[(data < -1) & (data > 1)] = np.nan

# Create a custom colormap using Turbo with red for the range -1 to 1
cmap = cm.get_cmap('turbo')
turbo_colors = cmap(np.linspace(0, 1, 253))
turbo_colors[126:128] = np.array([0, 0, 0, 1])  # Set the range -1 to 1 as grey
custom_cmap = colors.LinearSegmentedColormap.from_list('custom_cmap', turbo_colors)

im = ax.imshow(data, origin='upper', extent=[bounds.left, bounds.right, bounds.bottom, bounds.top],
               transform=ccrs.PlateCarree(), cmap=custom_cmap, vmin=-50, vmax=50)

# Add a colorbar with the updated range
cbar = fig.colorbar(im, ax=ax, orientation='horizontal', fraction=0.03, pad=0.05)
cbar.set_label('Tree Cover Difference %')

# Set the map title
ax.set_title('Global Tree Cover Difference (2020-2000)')

# Add map features like coastlines and borders
ax.coastlines(resolution='110m')

# Show the plot
plt.show()
```
