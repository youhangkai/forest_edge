```python
import os
import xarray as xr
import rasterio
from tqdm import tqdm
import numpy as np
directory = "G:/Hangkai/Global forest edge/2000Edge_classified"
output_nc = "G:/Global forest edge/2000Edge_classified/Edge_classified.nc"

# 获取所有的tif文件
tiff_files = [os.path.join(directory, f) for f in os.listdir(directory) if f.endswith('.tif')]

for idx, tiff in tqdm(enumerate(tiff_files), total=len(tiff_files), desc="Processing TIFFs"):
    with rasterio.open(tiff) as src:
        band1 = src.read(1)
        band2 = src.read(2)

        transform = src.transform
        x = np.arange(src.width) * transform.a + transform.c
        y = np.arange(src.height) * transform.e + transform.f

        da1 = xr.DataArray(band1, dims=('y', 'x'), coords={'y': y, 'x': x}, name='class')
        da2 = xr.DataArray(band2, dims=('y', 'x'), coords={'y': y, 'x': x}, name='diff')
        print(da1.coords)
        print(da2.coords)
        
"""
        if idx == 0:
            ds = xr.Dataset({'class': da1, 'diff': da2})
            ds.to_netcdf(output_nc, mode='w')
        else:
            with xr.open_dataset(output_nc) as existing_ds:
                # 使用merge合并现有的NetCDF和新的TIF数据
                merged_ds = xr.merge([existing_ds, xr.Dataset({'class': da1, 'diff': da2})])
                merged_ds.to_netcdf(output_nc, mode='w')  # 使用mode='w'以覆盖现有的NetCDF文件
"""
```
